```python
import numpy as np
import sys
sys.path.append('../wofs_phi')
import config as c
import utilities
from sklearn.metrics import brier_score_loss as BS
import os
from shutil import copy
import shutil
import copy
import matplotlib.pyplot as plt
import math
import pandas as pd
import matplotlib.ticker as ticker
import datetime as dt
#from wofs_phi import MLTrainer as t
#from wofs_phi import Grid
#import imageio.v3 as iio
#import skexplain
from copy import deepcopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import xarray as xr
import pickle
from matplotlib import rc, font_manager
from mpl_toolkits.axes_grid1 import make_axes_locatable
#from wofs.common.zarr import open_dataset
import matplotlib.colors as mpl_colors
import netCDF4 as ncdf
import statistics as stat
```

```python
xticks_font = font_manager.FontProperties(family='DejaVu Sans', style='normal',\
                                          size=12, weight='bold', stretch='normal')
yticks_font = font_manager.FontProperties(family='DejaVu Sans', style='normal',\
                                          size=12, weight='normal', stretch='normal')
legend_font = font_manager.FontProperties(family='DejaVu Sans', style='normal',\
                                          size=12, weight='normal', stretch='normal')

hazards = ['hail', 'wind', 'tornado']
leads = [0, 60, 120]
subplot_labels = np.array([['(a)', '(b)', '(c)'],\
                           ['(d)', '(e)', '(f)'],\
                           ['(g)', '(h)', '(i)']])

fig, axs = plt.subplots(len(hazards),len(leads))
fig.set_size_inches(16, 16)

save_dir = '/work/ryan.martz/wofs_phi_data/experiments/Paper_Figs/Correlations'
fig_save_file = 'probs_predictor_correlations.png'

for i in range(len(hazards)):
    
    hazard = hazards[i]
    
    for j in range(len(leads)):
        lead = leads[j]
        warning_directory = '/work/ryan.martz/wofs_phi_data/experiments/'\
        'wofs_psv3_with_torp_trained/warnings/length_60/'\
        'correlations/%s' %(hazard)
        warning_probs_cc_file = 'selected_predictor_%s_warning'\
        '_probs_correlations_%s-%smin.csv' %(hazard, lead+30, lead+90)
        
        lsr_directory = '/work/ryan.martz/wofs_phi_data/experiments/'\
        'wofs_psv3_with_torp_trained/obs/length_60/'\
        'correlations/%s' %(hazard)
        lsr_probs_cc_file = 'selected_predictor_%s_lsr'\
        '_probs_correlations_%s-%smin.csv' %(hazard, lead+30, lead+90)
        
        warning_probs_ccs_df = pd.read_csv('%s/%s' %(warning_directory, warning_probs_cc_file))
        lsr_probs_ccs_df = pd.read_csv('%s/%s' %(lsr_directory, lsr_probs_cc_file))
        
        warning_probs_ccs = np.array(warning_probs_ccs_df['Correlation'])
        lsr_probs_ccs = np.array(lsr_probs_ccs_df['Correlation'])
        plvs = np.array(warning_probs_ccs_df['Predictor'])
        
        #keep_indices = np.where(plvs != 'WoFS Initialization Time')
        #warning_probs_ccs = warning_probs_ccs[keep_indices]
        #lsr_probs_ccs = lsr_probs_ccs[keep_indices]
        #plvs = plvs[keep_indices]
        
        ps_plvs = []
        ps_plv_indices = []
        wofs_plvs = []
        wofs_plv_indices = []
        torp_plvs = []
        torp_plv_indices = []
        loc_plvs = []
        loc_plv_indices = []
        for k in range(len(plvs)):
            plv = plvs[k]
            if '30km' in plv:
                plv = plv[:-7]
            plvs[k] = plv
            if 'PS Prob' in plv:
                ps_plvs.append(plv)
                ps_plv_indices.append(k)
            elif plv == 'Latitude' or plv == 'Longitude':
                loc_plvs.append(plv)
                loc_plv_indices.append(k)
            elif plv == 'WoFS Initialization Time':
                loc_plvs.append('WoFS Init Time')
                loc_plv_indices.append(k)
            elif 'TORP' in plv:
                torp_plvs.append(plv)
                torp_plv_indices.append(k)
            else:
                wofs_plvs.append(plv)
                wofs_plv_indices.append(k)
        
        ordered_plvs = []
        colors = []
        for k in range(len(ps_plvs)):
            ordered_plvs.append(ps_plvs[k])
            colors.append('darkmagenta')
        for k in range(len(wofs_plvs)):
            ordered_plvs.append(wofs_plvs[k])
            colors.append('chocolate')
        for k in range(len(torp_plvs)):
            ordered_plvs.append(torp_plvs[k])
            colors.append('deepskyblue')
        for k in range(len(loc_plvs)):
            ordered_plvs.append(loc_plvs[k])
            colors.append('k')
        
        plvs = ordered_plvs
        order_indices = ps_plv_indices
        order_indices.extend(wofs_plv_indices)
        order_indices.extend(torp_plv_indices)
        order_indices.extend(loc_plv_indices)
        warning_probs_ccs = warning_probs_ccs[order_indices]
        lsr_probs_ccs = lsr_probs_ccs[order_indices]
        
        if j == 0:
            axs[i,j].bar(np.arange(len(plvs))-0.15, warning_probs_ccs, width = 0.3, color = 'b', label = 'Warnings Trained')
            axs[i,j].bar(np.arange(len(plvs))+0.15, lsr_probs_ccs, width = 0.3, color = 'y', label = 'Reports Trained')
        else:
            axs[i,j].bar(np.arange(len(plvs))-0.15, warning_probs_ccs, width = 0.3, color = 'b')
            axs[i,j].bar(np.arange(len(plvs))+0.15, lsr_probs_ccs, width = 0.3, color = 'y')
            #axs[i,j].plot(np.arange(len(plvs)), warning_probs_ccs, color = 'b', linewidth = 3)
            #axs[i,j].plot(np.arange(len(plvs)), lsr_probs_ccs, color = 'r', linewidth = 3)
        
        axs[i,j].set_xticks(np.arange(len(plvs)))
        axs[i,j].set_yticks(np.arange(0.1, 0.95, 0.1))
        
        
        if j == 0:
            for label in axs[i,0].get_yticklabels():
                label.set_fontproperties(yticks_font)
        else:
            axs[i,j].yaxis.set_ticklabels([])
        
        if i == 2:
            axs[i,j].set_xticklabels(plvs, rotation = 90)
            for m in range(len(axs[i,j].get_xticklabels())):
                label = axs[i,j].get_xticklabels()[m]
                label.set_fontproperties(xticks_font)
                label.set_color(colors[m])
        else:
            axs[i,j].xaxis.set_ticklabels([])
        
        axs[i,j].set_title('%s-%s Minutes' %(lead, lead+60), fontname = 'DejaVu Sans',\
                         fontsize = 12, fontweight = 'bold')
        axs[i,j].set_ylim([0,0.92])
        axs[i,j].text(0, 1.02, subplot_labels[i,j], weight = 'bold', fontsize = 14,\
                   transform=axs[i,j].transAxes)
        axs[i,j].spines["top"].set_visible(True)
        axs[i,j].spines["right"].set_visible(True)
        axs[i,0].set_ylabel(hazard.capitalize(), fontsize = 14, fontweight = 'bold', fontname = 'DejaVu Sans')
        
        
    axs[0,0].legend(prop = legend_font, loc = 'upper right')

    fig.supylabel('Absolute Correlation Coefficient', fontsize = 18, fontweight = 'bold', fontname = 'DejaVu Sans')
    #fig.suptitle(hazard.capitalize(), fontsize = 18, fontweight = 'bold', fontname = 'DejaVu Sans')
    fig.tight_layout()
    fig.savefig('%s/%s' %(save_dir, fig_save_file), dpi=300)
```
